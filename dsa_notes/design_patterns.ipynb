### Creational Design Pattern

```python
# Singleton
class Singleton:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance
```

```python
# Factory
from abc import ABC, abstractmethod
# abstract product
class Shape(ABC):
    @abstractmethod
    def draw(self):
        pass
# concrete Product
class Circle(Shape):
    def draw(self):
        print("Drawing a Circle")

class Triangle(Shape):
    def draw(self):
        print("Drawing a Triangle")

# Factory
class ShapeFactory:
    @staticmethod
    def create_shape(shape_type):
        if shape_type == "circle":
            return Circle()
        elif shape_type == "triangle":
            return Triangle()
        else:
            raise Exception("Invalid Shape")
# Client Code
factory = ShapeFactory()
shape1 = factory.create_shape("circle")
shape1.draw()

shape2 = factory.create_shape("triangle")
shape2.draw()
```

```python
# Decorator pattern
def uppercase_decorator(func):
    def wrapper():
        return func().upper()
    return wrapper

@uppercase_decorator
def say_hello():
    return "hello World!"

print(say_hello())
print(say_hello.__name__)
```

```python
import functools

def log_decorator(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        # do you thing here
        print(f"Calling function {func.__name__}")
        print(f"Postional Arguments: {args}")
        print(f"Keyword Arguments: {kwargs}")

        result = func(*args, **kwargs)

        print(f"Returned : {result}")
        return result
    return wrapper

@log_decorator
def add(a, b):
    return a + b

def greet(name, greetings="Bonjour"):
    return f"{greetings} {name}"

result_add = add(3, 5)
print(add.__name__)
```

```python
# Adapter Design Pattern
# The Target: The interface the client expects
class Target:
    def request(self):
        """The client expects this method signature."""
        return "Target: Standard request."

# The Adaptee: The class with the incompatible interface
class ThirdPartyService:
    def specific_request(self):
        """The Adaptee has this different method signature."""
        return "Adaptee: Specific request processed."

# the adapter 
class Adapter(Target):
    def __init__(self, adaptee = ThirdPartyService):
        self._adaptee = adaptee
    def request(self):
        # translating the Info         
        return self._adaptee.specific_request()
```

### Behavioural Design Patterns

```python
# Iterator Pattern
class CountIterator:
    def __init__(self, start, end):
        self._current = start
        self._end = end

    def __iter__(self):
        # Returns the iterator object itself
        return self

    def __next__(self):
        if self._current <= self._end:
            value = self._current
            self._current += 1  # Update state/position
            return value
        else:
            # Signals the end of the iteration
            raise StopIteration
```
