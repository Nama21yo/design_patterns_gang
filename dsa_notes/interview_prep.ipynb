## 📝 Difference Between Shallow Copy and Deep Copy

The distinction between shallow and deep copying matters when you are copying a **compound object** (an object that contains other objects, like a list of lists or a dictionary of lists).

Both methods are available in Python via the `copy` module (`copy.copy()` for shallow, `copy.deepcopy()` for deep).

### 1. Shallow Copy (`copy.copy()`)

A shallow copy creates a new compound object, but it inserts **references** to the objects found in the original.

* **New Object:** Yes, the top-level container (e.g., the outer list or dictionary) is new.
* **Contained Objects:** The objects *inside* the container are **not** copied; their memory addresses are simply referenced by the new copy.
* **Result:** Changes made to a *mutable* element in the copy (e.g., changing a nested list) will also affect the original, because they point to the same memory location.

**Use Case:** When speed is critical, and the container only holds immutable objects (like integers or strings), or if you intentionally want the copies to share internal mutable objects.



### 2. Deep Copy (`copy.deepcopy()`)

A deep copy creates a new compound object and then recursively inserts **copies** of the objects found in the original.

* **New Object:** Yes, the top-level container is new.
* **Contained Objects:** The objects *inside* the container are also copied, recursively. They are new objects residing in a different memory location.
* **Result:** Changes made to any element (even nested mutable ones) in the deep copy will **not** affect the original, ensuring complete independence.

**Use Case:** When dealing with complex, nested mutable structures (lists of lists, dictionaries containing objects) where you need the copy to be fully independent of the original.



| Feature | Shallow Copy (`copy.copy()`) | Deep Copy (`copy.deepcopy()`) |
| :--- | :--- | :--- |
| **Top-Level** | New Object | New Object |
| **Nested Objects** | References Original Objects | Creates Independent Copies Recursively |
| **Memory** | Shared for nested mutable items | Fully Independent |
| **Performance** | Faster | Slower (due to recursion) |
| **Independence**| Low (changes affect the original) | High (fully isolated) |

### 1. 🔑 Hashing & Data Structures

* **Core Idea of Hashing:** Mapping a large key to a small array index using a deterministic **hash function** to achieve $O(1)$ average-case performance.
* **Data Structures:** A hash map uses an underlying array (bucket array) and a mechanism to store key-value pairs.
* **Collision Resolution:** Methods to handle multiple keys mapping to the same index: **Separate Chaining** (linked lists in buckets) and **Open Addressing/Probing** (finding the next empty slot).
* **Keys and Hashability:** Keys must be **hashable** (immutable) to guarantee a consistent hash value throughout their lifetime. Custom objects must implement `__hash__` and `__eq__`.
* **Python Implementations:** The use of `dict` (hash map) and `set` (hash set), noting that modern Python dictionaries preserve insertion order.
* **`HashMap` vs. `Hashtable`:** The difference primarily lies in the legacy `Hashtable` being **synchronized** (thread-safe but slower) versus the modern `HashMap` being non-synchronized and faster.

### 2. 🐍 Python Utilities

* **`defaultdict`:** A specialized dictionary that simplifies grouping and counting by automatically providing a default value (via the `default_factory`) for any accessed missing key, eliminating the need for explicit `if key not in dict:` checks.
* **Dictionary Methods:** Overview of methods like `d[key]`, `d.get()`, `d.update()`, `d.pop()`, and views (`d.keys()`, `d.values()`, `d.items()`).

### 3. 🧵 Concurrency & Memory

* **Threads:** Lightweight units of execution that **share the memory space** of their parent process, making communication fast but creating potential for **race conditions**.
* **Race Conditions:** Occur when multiple threads access shared, mutable data, and the outcome depends on the unpredictable order of execution (e.g., the read-increment-write problem).
* **Thread Safety:** Achieved using synchronization primitives like **Locks** (Mutexes) to ensure that the critical section of code runs **atomically** (by only one thread at a time).

Designing a HashMap requires understanding how to map an infinite space of keys to a finite array while handling the inevitable collisions efficiently.

Here is the internal design breakdown, addressing your specific questions about SHA-256, collision handling, and dynamic resizing.

## 1\. The Core Architecture

At its heart, a HashTable is indeed just an **Array**.

  * **The Array (Buckets):** We allocate a fixed-size array (say, size $M=16$ initially). Each slot in this array is often called a "bucket".
  * **The Mapping:** We need a way to determine which bucket a specific key (like "username") belongs to.

### The Hash Function

The hash function is the magic that converts a key into an array index.

$$\text{Index} = \text{hash(key)} \ \% \ \text{ArraySize}$$

1.  **`hash(key)`**: Converts the object into a raw integer (the hash code).
2.  **`% ArraySize`**: The modulo operator ensures the integer fits within the bounds of our array indices ($0$ to $M-1$).

### ❓ Is SHA-256 a Hash Function?

**Yes**, SHA-256 is a hash function, but it is a **Cryptographic Hash Function**.

  * **SHA-256:** Designed for security. It is intentionally complex to prevent reverse-engineering and collisions. It returns a massive 256-bit number. It is **too slow** for use in a standard HashMap data structure.
  * **HashMap Hash Functions:** Designed for **speed** and **uniform distribution**. Python's internal `hash()` or Java's `hashCode()` are non-cryptographic and optimized for fast lookups.

-----

## 2\. Collision Handling Schemes

Since our array size is small compared to the number of possible keys, two different keys will eventually hash to the same index. This is a **Collision**.

### A. Separate Chaining (Most Common)

Each bucket in the array actually stores a pointer to a **Linked List** (or a Balanced BST in Java 8+).

  * **Insert:** Calculate index. If the bucket is empty, add the node. If occupied, traverse the list to check if key exists (update it) or append to the end.
  * **Search:** Calculate index. Traverse the linked list at that index.
  * **Delete:** Calculate index. Traverse the list and remove the node.

### B. Open Addressing (Probing)

We do not use an extra data structure. Everything is stored in the array itself. If `index` is taken, we find another spot.

  * **Linear Probing:** Check `index + 1`, then `index + 2`, etc.
  * **Quadratic Probing:** Check `index + 1^2`, `index + 2^2`, etc.
  * **Drawback:** Suffers from "clustering" (consecutive occupied slots), which slows down lookups.

-----

## 3\. Dynamic Resizing & Load Factor

This is where the **"Amortized $O(1)$"** comes from.

### The Load Factor ($\alpha$)

The load factor is a measure of how full the hash table is:
$$\alpha = \frac{\text{Number of Elements (N)}}{\text{Size of Array (M)}}$$

  * **Threshold:** A common default is **0.75**.
  * **Why?**
      * If $\alpha$ is too high (e.g., 0.9), collisions increase, and lists get long. Search degrades to $O(N)$.
      * If $\alpha$ is too low (e.g., 0.1), we waste memory.

### The Rehash Process

When `count / size > 0.75`, we must resize.

1.  **Double the Capacity:** Create a new array with size $2 \times M$.
2.  **Rehash EVERYTHING:** We cannot just copy elements over.
      * Why? Because the index depends on the array size (`hash % size`). If size changes, the index changes.
      * *Example:* Key `10` in size `5` is at index `0`. In size `10`, it is at index `0`. Key `7` in size `5` is at index `2`. In size `10`, it is at index `7`.
3.  **Cost:** This operation takes $O(N)$ time.

**Why Amortized $O(1)$?**
Although resizing takes $O(N)$, it happens very rarely (doubling size means the next resize is far away). When you average the cost over all insertions, it is constant time.

## 4\. Python Implementation (Design HashMap)

Here is a clean implementation using **Separate Chaining** with a basic Python list acting as the Linked List.

```python
class ListNode:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None

class MyHashMap:
    def __init__(self):
        # Initial capacity
        self.capacity = 1000 
        self.buckets = [None] * self.capacity
        
    def _hash(self, key):
        # Python's built-in hash() handles strings, ints, etc.
        return hash(key) % self.capacity

    def put(self, key, value):
        index = self._hash(key)
        
        # If bucket is empty, create new node
        if self.buckets[index] is None:
            self.buckets[index] = ListNode(key, value)
            return
        
        # Collision: Traverse the linked list
        curr = self.buckets[index]
        while True:
            # Case 1: Key already exists -> Update value
            if curr.key == key:
                curr.value = value
                return
            
            # Case 2: Reached end of list -> Append new node
            if curr.next is None:
                curr.next = ListNode(key, value)
                return
            
            curr = curr.next

    def get(self, key):
        index = self._hash(key)
        curr = self.buckets[index]
        
        while curr:
            if curr.key == key:
                return curr.value
            curr = curr.next
            
        return -1 # Key not found

    def remove(self, key):
        index = self._hash(key)
        curr = self.buckets[index]
        prev = None
        
        while curr:
            if curr.key == key:
                if prev:
                    prev.next = curr.next # Bypass current node
                else:
                    self.buckets[index] = curr.next # Update head
                return
            prev = curr
            curr = curr.next
```

### Complexity Analysis

| Operation | Average Case | Worst Case | Why Worst Case? |
| :--- | :--- | :--- | :--- |
| **Insert** | $O(1)$ | $O(N)$ | All keys collide into one bucket (linked list). |
| **Search** | $O(1)$ | $O(N)$ | Traversing a long linked list. |
| **Delete** | $O(1)$ | $O(N)$ | Finding the node in a long list. |

**Note:** In modern implementations (like Java 8+), the "Worst Case" is improved to $O(\log N)$ by converting long linked lists into Balanced Binary Search Trees (Red-Black Trees) once a bucket exceeds a certain size (usually 8 nodes).

### Longest Consecutive Sequence


````python
# Problem: Find length of sequence (e.g., 1, 2, 3, 4) in an unsorted array. 
# Input: [100, 4, 200, 1, 3, 2] Output: 4 (1, 2, 3, 4)

def longest_consecutive(nums):
    num_set = set(nums)
    longest = 0

    for num in num_set:
        # check if it is a start sequence
        if (num - 1) not in num_set:
            current_num = num
            current_long = 1

            while (current_num + 1) in num_set:
                current_long += 1
                current_num += 1
            longest = max(current_long, longest)
    return longest

print(longest_consecutive([100, 4, 200, 1, 3, 2]))
````

````python
def is_permutation(s1, s2):
    if len(s1) != len(s2):
        return False

    counts = [0] * 26

    for char in s1:
        counts[ord(char) - ord("a")] += 1

    for char in s2:
        index = ord(char) - ord("a")
        counts[index] -= 1
        if counts[index] < 0:
            return False
    return True
````

````python
def find_anagrams(s, p):
    if len(p) > len(s):
        return []

    pCount, sCount = {}, {}
    # build frequenct map
    for i in range(len(p)):
        pCount[p[i]] = 1 + pCount.get(p[i], 0)
        sCount[s[i]] = 1 + sCount.get(s[i], 0)

    res = [0] if sCount == pCount else []

    left = 0
    for right in range(len(p), len(s)):
        # add new char -> right
        sCount[s[right]] = 1 + sCount.get(s[right], 0)

        # remove the old char -> left
        sCount[s[left]] -= 1

        if sCount[s[left]] == 0:
            sCount.pop(s[left])

        left += 1
        if sCount == pCount:
            res.append(left)

    return res
````

### Design Ordered Stream

````python
class OrderedStream:

    def __init__(self, n: int):
        self.ptr = 1
        self.data = [None] * (n + 1)

    def insert(self, idKey: int, value: str):
        self.data[idKey] = value
        ans = []
        while self.ptr < len(self.data) and self.data[self.ptr]:
            ans.append(self.data[self.ptr])
            self.ptr += 1
        return ans


# Your OrderedStream object will be instantiated and called as such:
# obj = OrderedStream(n)
# param_1 = obj.insert(idKey,value)

# Time Complexity: $O(1)$ Amortized.
# Space Complexity: $O(N)$.
````

## Storing Graphs
There are two primary ways to store a graph. The choice depends on the graph's density (ratio of edges to nodes).

# Graph Traversals
### 1. Breadth-First Search (BFS)
* **Strategy:** Explore layer by layer (neighbors first, then neighbors of neighbors).
* **Data Structure:** Queue (`deque`).
* **Use Case:** Finding the **Shortest Path** in an **unweighted** graph.

### 2. Depth-First Search (DFS)
* **Strategy:** Go deep into one branch before backtracking.
* **Data Structure:** Stack (or Recursion).
* **Use Case:** Detecting cycles, topological sorting, puzzle solving (mazes).

**Crucial Note:** You MUST use a `visited` set to prevent infinite loops in cyclic graphs.


# Question 1: Number of Connected Components
**Problem:** Given an undirected graph, find the number of distinct connected sub-graphs.

**Idea:**

1. Initialize a `count = 0`.
2. Iterate through all nodes from `0` to `n`.
3. If a node is **not visited**, it belongs to a new component.
* Increment `count`.
* Launch a traversal (DFS or BFS) to find and mark *all* nodes connected to this node as visited.


4. Return `count`.

**Complexity:**

* **Time:** O(V + E) (We visit every node and edge once).
* **Space:** O(V) (Visited set + Recursion stack).

```python
def count_components(n, edges):
    # 1. Build Graph
    adj = defaultdict(list)
    for u, v in edges:
        adj[u].append(v)
        adj[v].append(u)
    
    visited = set()
    count = 0
    
    # 2. Define Traversal (DFS)
    def dfs(node):
        for neighbor in adj[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                dfs(neighbor)

    # 3. Iterate all nodes
    for i in range(n):
        if i not in visited:
            count += 1
            visited.add(i)
            dfs(i) # This call marks the entire component as visited
            
    return count

```

# Question 2: Number of Islands
**Problem:** Given a grid of `'1'`s (land) and `'0'`s (water), count the number of islands. An island is surrounded by water and is formed by connecting adjacent lands horizontally or vertically (4-way).

**Approach: In-Place Modification**
Instead of using a separate `visited` set (which uses O(M \times N) extra space), we can modify the input grid itself. When we visit a land cell `'1'`, we change it to `'0'` (or `'#'`) to mark it as processed. This effectively "sinks" the island.

### The Algorithm
1. Iterate through every cell `(r, c)` in the grid.
2. If the cell is `'1'`:
* We found a **new island**. Increment `count`.
* Call `DFS(r, c)` to sink the entire island (turn all connected `'1'`s to `'0'`s).



```python
def num_islands(grid):
    if not grid:
        return 0
    
    rows, cols = len(grid), len(grid[0])
    count = 0
    
    def dfs_sink(r, c):
        # Base Case: Check bounds and if it is water
        if r < 0 or c < 0 or r >= rows or c >= cols or grid[r][c] == "0":
            return
        
        # Mark as visited (Sink the land)
        grid[r][c] = "0"
        
        # Visit all 4 neighbors
        dfs_sink(r + 1, c) # Down
        dfs_sink(r - 1, c) # Up
        dfs_sink(r, c + 1) # Right
        dfs_sink(r, c - 1) # Left

    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == "1":
                count += 1
                dfs_sink(r, c) # Sinks the whole island found at (r,c)
                
    return count

# Complexity Analysis
# Time: O(M * N) - We visit every cell at least once.
# Space: O(M * N) - In worst case (grid is all land), recursion stack is M*N.

```

### The "Why" behind the Output ChangeYou asked why the matrix changes.

* **Before:** The grid contains `'1'`s representing land.
* **During:** As the algorithm finds an island, it traverses every connected `'1'` and flips it to `'0'`.
* **After:** The grid becomes all `'0'`s (all water) because every island has been "sunk" (processed). This mutation allows us to avoid recounting the same land cells without using extra memory for a `visited` set.

````python
from collections import deque

class Solution:
    def orangesRotting(self, grid):
        if not grid:
            return 0
            
        rows, cols = len(grid), len(grid[0])
        queue = deque()
        
        # 1. distance[i][j] stores the time (minute) it took for cell (i,j) to rot
        distance = [[float("inf") for _ in range(cols)] for _ in range(rows)]
        fresh_count = 0
        
        # 2. Initial Scan
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == 2:
                    queue.append((r, c))
                    distance[r][c] = 0  # Sources start at minute 0
                elif grid[r][c] == 1:
                    fresh_count += 1
        
        # Edge Case: If there are no fresh oranges, it's already "done" in 0 mins
        if fresh_count == 0:
            return 0
            
        directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        
        # 3. Multi-Source BFS
        while queue:
            r, c = queue.popleft()
            
            for dr, dc in directions:
                nr, nc = r + dr, c + dc
                
                # Check: Inbound, Unvisited (inf), and is a Fresh Orange
                if (0 <= nr < rows and 0 <= nc < cols and 
                    distance[nr][nc] == float("inf") and 
                    grid[nr][nc] == 1):
                    
                    # Rot the orange
                    grid[nr][nc] = 2
                    fresh_count -= 1
                    
                    # The minute it rots is the previous orange's minute + 1
                    distance[nr][nc] = distance[r][c] + 1
                    queue.append((nr, nc))
        
        # 4. Final Result Calculation
        if fresh_count > 0:
            return -1 # Some oranges were unreachable
            
        # The answer is the highest number in our distance matrix
        max_minutes = 0
        for r in range(rows):
            for c in range(cols):
                if distance[r][c] != float("inf"):
                    max_minutes = max(max_minutes, distance[r][c])
                    
        return max_minutes
        
# Optimal Inplace
def orangeRotting(grid):
    if not grid:
        return 0

    rows, cols = len(grid), len(grid[0])

    queue = deque()
    fresh_count = 0

    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == 2:
                queue.append((r, c))
            elif grid[r][c] == 1:
                fresh_count += 1

    if fresh_count == 0:
        return 0

    minutes = 0
    directions = [(0, 1), (0 , -1), (1, 0), (-1 , 0)]

    while queue and fresh_count > 0:

        for _ in range(len(queue)):
            r, c = queue.popleft()

            for dr, dc in directions:
                nr, nc = r + dr, c + dc

                if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] == 1:
                    grid[nr][nc] = 2
                    fresh_count -= 1
                    queue.append((nr, nc))
        minutes += 1

    return minutes if fresh_count == 0 else -1
````

### https://leetcode.com/problems/lru-cache

````python
# A Cache is a high-speed data storage layer that stores a subset of data so that 
# future requests for that data are served faster than accessing the data's
# primary storage location. It trades capacity (it is small) for speed (it is fast).

class Node:
    def __init__(self, key, value):
        self.key = key 
        self.value = value
        self.prev = None
        self.next = None

class LRUCache:
    def __init__(self, capacity):
        self.capacity = capacity
        self.cache = {} # hash map

        # head -> MRU -> ... -> LRU -> tail
        self.head = Node(0, 0)
        self.tail = Node(0, 0)
        self.head.next = self.tail
        self.tail.prev = self.head

    def _remove(self, node):
        prev_node = node.prev
        next_node = node.next
        prev_node.next = next_node
        next_node.prev = prev_node

    def _add_to_head(self, head):
        node.prev = self.head
        node.next = self.head.next
        self.head.next.prev = node
        self.head.next = node
        
    def get(self, key):
        if key in self.cache:
            node = self.cache[key]

            # remove from the current and move to the head
            self._remove(node)
            self._add_to_head(node)
            return node.value

        return -1

    def put(self, key, value):
        if key in self.cache:
            node = self.cache[key]
            node.value = value
            # remove from the current and move to the head
            self._remove(node)
            self._add_to_head(node)
        else:
            # add new key
            new_node = Node(key, value)
            self.cache[key] = new_node
            self._add_to_head(new_node)

            # check capacity and evict
            if len(self.cache) > self.capacity:
                lru_node = self.tail.prev
                self._remove(lru_node)
                del self.cache[lru_node.key] # we should also remove from the map
````

````python
# Part-1
# Problem: Given an array of meeting time intervals [[s1,e1],[s2,e2],...],
# determine if a person could attend all meetings. Translation: Do any of the intervals overlap?

def canAttendMeetings(intervals):
    intervals.sort(key= lambda x: x[0])

    for i in range(1, len(intervals)):
        curr_start = intervals[i][0]
        prev_end = intervals[i - 1][1]

        if prev_end > curr_start:
            return False

    return True
# Time: O(N log N)
# Space: O(1)

# Part-2
# Problem: Given the same intervals, find the minimum number of conference rooms required. 
    # Translation: What is the maximum number of meetings happening simultaneously?

def minMeetingRooms_two_pointers(intervals):
    if not intervals:
        return 0
    
    start_time = sorted([i[0] for i  in intervals])
    end_time = sorted([i[1] for i in intervals])

    # left = 0
    # right = 0
    start = end = 0
    res = count = 0

    while start < len(intervals):

        if start_time[start] < end_time[end]:
            count += 1
            start += 1
        else:
            count -= 1
            end += 1
        res = max(count, res)

    return res

# Time Complexity: O(N log N) due to sorting.
# Space Complexity: O(N) to store the start/end arrays.


# Method-2: Line Sweeps with Prefix Sum
# size = max of the end time
# [start, end]
# [[0,5],[5,6],[2,4]]

# [1, 0, 0, 0, -1, 0, 0]
# let us do the prefix sum
# [1, 1, 1, 1, 0, 0, 0] -> meeting started at 0 and ended at 4
# for the next part
# [1, 0, 0, 0, 0, -1]
# for the last part
# [1, 0, 1, 0, -1, -1]
# [1, 1, 2, 2, 1, 0]
def minMeetingRooms_lineSweep(intervals):
    max_time = 0

    for s,e in intervals:
        max_time = max(max_time, e)

    timeline = [0] * (max_time + 1)

    for s,e in intervals:
        timeline[s] += 1
        timeline[e] -= 1

    max_rooms = 0
    current_rooms = 0

    for i in range(max_time + 1):
        current_rooms += timeline[i]
        # if current_rooms > max_rooms:
        #     max_rooms = current_rooms
        max_rooms = max(max_rooms, current_rooms)

    return max_rooms
# Time Complexity: O(N + K) where N is intervals and K is the max_time.
# Space Complexity: O(K) for the array.
````

### Two City Scheduling

````python
class Solution:
    def twoCitySchedCost_BruteForce(self, costs):
        n = len(costs) // 2
        
        def solve(index, countA, countB):
            # Base Case: All people processed
            if index == len(costs):
                return 0
            
            cost = float('inf')
            
            # Option 1: Fly to A (if spots available)
            if countA < n:
                cost = min(cost, costs[index][0] + solve(index + 1, countA + 1, countB))
                
            # Option 2: Fly to B (if spots available)
            if countB < n:
                cost = min(cost, costs[index][1] + solve(index + 1, countA, countB + 1))
                
            return cost

        return solve(0, 0, 0)

# Time Complexity: O(2^N) - Exponential, very slow.
# Space Complexity: O(N) for recursion stack.

class Solution:
    def twoCitySchedCost_DP(self, costs):
        n = len(costs) // 2
        memo = {}
        
        def solve(index, countA):
            if index == len(costs):
                return 0
            
            state = (index, countA)
            if state in memo:
                return memo[state]
            
            countB = index - countA
            min_cost = float('inf')
            
            # Try A
            if countA < n:
                min_cost = min(min_cost, costs[index][0] + solve(index + 1, countA + 1))
            
            # Try B
            if countB < n:
                min_cost = min(min_cost, costs[index][1] + solve(index + 1, countA))
            
            memo[state] = min_cost
            return min_cost

        return solve(0, 0)

# Time Complexity: O(N^2) - There are N indices and N possible values for countA.
# Space Complexity: O(N^2) for the memoization table.


# optimal - nlogn
def twoCityScheduling(costs):
    diffs = []
    for a, b in costs:
        diffs.append([b - a, a, b])

    diffs.sort(key= lambda x : x[0])
    res = 0
    mid = len(diffs) // 2
    for i in range(len(diffs)):
        if i < mid:
            res += diffs[i][2]
        else:
            res += diffs[i][1]
    return res

# Follow-up: what if we have odd number of people rather than even
class Solution:
    def minCostOddOrEven(self, costs):
        # 1. Sort by Opportunity Cost (costA - costB)
        costs.sort(key=lambda x: x[0] - x[1])
        
        n = len(costs)
        mid = n // 2
        
        # Helper to calculate cost given a split index
        # People before split -> A, People at and after split -> B
        def calculate_split_cost(split_index):
            current_total = 0
            for i in range(n):
                if i < split_index:
                    current_total += costs[i][0] # Go to A
                else:
                    current_total += costs[i][1] # Go to B
            return current_total

        # Case 1: Even number of people (Standard LeetCode)
        if n % 2 == 0:
            return calculate_split_cost(mid)
            
        # Case 2: Odd number of people (Bloomberg Twist)
        else:
            # Option A: City A gets the extra person (A has mid+1, B has mid)
            # Split happens at index (mid + 1)
            cost_A_gets_extra = calculate_split_cost(mid + 1)
            
            # Option B: City B gets the extra person (A has mid, B has mid+1)
            # Split happens at index (mid)
            cost_B_gets_extra = calculate_split_cost(mid)
            
            return min(cost_A_gets_extra, cost_B_gets_extra)

# Example Usage:
# costs = [[10,20], [30,200], [400,50], [30,20], [10, 20]] (5 people)
# solver = Solution()
# print(solver.minCostOddOrEven(costs))
````

````python
# Validate BST
# Given the root of a binary tree, determine if it is a valid binary search tree (BST).

# A valid BST is defined as follows:

# The left subtree of a node contains only nodes with keys strictly less than the node's key.
# The right subtree of a node contains only nodes with keys strictly greater than the node's key.
# Both the left and right subtrees must also be binary search trees.

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isValidBST(self, root):
        def checkbst(root, mini, maxi):
            if root is None:
                return True
            return mini <= root.val and root.val <= maxi and checkbst(root.left, mini, root.val - 1) and checkbst(root.right, root.val + 1, maxi)
        return checkbst(root, float("-inf"), float("inf"))

def validateBST(root):
    def checkBST(root, low, high):
        if root is None:
            return True

        if not (low <= root.val <= high):
            return False

        return checkBST(root.left, low, root.val) and checkBST(root.right, root.val, high)
    return checkBST(root, float("-inf"), float("inf"))

# Time Complexity: $O(N)$ (We visit every node once).
# Space Complexity: $O(H)$ (Recursion stack depth, where $H$ is height).


# Approach-2
# Key Insight: If you perform an Inorder Traversal (Left -> Root -> Right) on a valid BST
# , the resulting list of values will always be strictly sorted in ascending order.

def isValidBST(root):
    if not root:
        return True

    queue = deque([root, -math.inf, math.inf])

    while queue:
        node, low, high = queue.popleft()

        if not (low < node.val < high):
            return False

        if node.left:
            queue.append((node.left, low, node.val))

        if node.right:
            queue.append((node.right, node.val, high))

    return True
````

````python
# Example: s = "abbaca" -> ca
def remove_duplicates(s):
    stack = []
    for char in s:
        if stack and stack[-1] == char:
            stack.pop()
        else:
            stack.append(char)

    return "".join(stack)
# time and space O(N)

# Example: s = "deeedbbcccbdaa", k = 3
# eee removes -> ddbbcccbdaa

# ccc removes -> ddbbbdaa

# bbb removes -> dddaa

# ddd removes -> aa Result: "aa"

def removeDuplicates(s, k):
    # stack stores [char, count]
    stack = []

    for char in s:
        if stack and stack[-1][0] == char:
            stack[-1][1] += 1 # same character increment count

            if stack[-1][1] == k:
                stack.pop()
        else:
            stack.append([char, 1])

    # construct the string
    result = []
    for char, count in stack:
        result.append(char * count)

    return "".join(result)
````

### Edit Distance

````python
# Input: word1 = "horse", word2 = "ros"

# Output: 3

# horse -> rorse (Replace 'h' with 'r')

# rorse -> rose (Delete 'r')

# rose -> ros (Delete 'e')

def minDistance(word1, word2):
    m, n = len(word1), len(word2)

    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m + 1):
        dp[i][0] = i

    for j in range(n + 1):
        dp[0][j] = j


    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i - 1] == word2[j - 1]:
                # character matches take from the diagonal
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = 1 + min(dp[i][j - 1], # insert
                                   dp[i - 1][j], # delete
                                   dp[i - 1][j - 1] # replace
                                  )
    return dp[m][n]
# time and space: O(M*N)
````

### https://leetcode.com/problems/evaluate-division

````python
# The Approach: Graph Build + DFS/BFS
# {
#   "a": {"b": 2.0},
#   "b": {"a": 0.5}
# }

from collections import defaultdict
def calcEquation(equations, values, queries):
    graph = defaultdict(dict)

    for (start, end), value in zip(equations, values):
        graph[start][end] = value
        graph[end][start] = 1 / value

    def dfs(curr_node, target, visited):
        # if we found the target
        # the multiplier contribution is 1.0
        if curr_node == target:
            return 1.0

        visited.add(curr_node)

        for neigh, weight in graph[curr_node].items():
            if neigh not in visited:
                result = dfs(neigh, target, visited)

                if result != -1.0:
                    return weight * result
        return -1.0

    results = []
    for src, dst in queries:
        if src not in graph or dst not in graph:
            results.append(-1.0)
        elif src == dst:
            results.append(1.0)
        else:
            results.append(dfs(src, dst, set()))
    return results
````
